```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Arial"
```

```python
Data=pd.read_csv("Correlation-RoF.csv")
```

```python
boxprops=dict(linestyle='-', linewidth=4,  color='Blue')
medianprops = dict(linestyle='-', linewidth=4, color='green')
whiskerprops=dict(linestyle='-', linewidth=4, color='Blue')
capprops=dict(linestyle='-', linewidth=4, color='black')
flierprops = dict(marker='o', markersize=18,
                  linestyle='none', markeredgecolor="black")
```

```python
figure=Data.boxplot(rot=45,figsize=(90,45),fontsize=70,boxprops=boxprops, medianprops=medianprops,whiskerprops=whiskerprops, capprops=capprops,  grid=True,flierprops=flierprops)
                    #flierprops=flierprops
figure.set_ylabel("correlation coefficient",fontsize=70)
figure.set_ylim(0,1)
```

```python
figure2=Data.plot.box( rot=40,figsize=(20,15),fontsize=20, patch_artist = True, color='green')
figure2.set_ylabel("correlation coefficient",fontsize=14)
```
